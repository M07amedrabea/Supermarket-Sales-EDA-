```
data.sample(1)
```

```
data['Payment'].value_counts()
```

```
data['Payment'].value_counts().plot(kind='bar')
plt.show()
```

```
data['Branch'].value_counts()
```

```
data['Branch'].value_counts().plot(kind='bar')
plt.show()
```

```
data['City'].value_counts().plot(kind='bar')
plt.show()
```

```
data['Customer type'].value_counts().plot(kind='bar')
plt.show()
```

```
data['Product line'].value_counts().plot(kind='bar')
plt.show()
```

# Line Plot

```
plt.figure(figsize=(10,5))
#fig , ax = plt.subplots(figsize=(10,5))
data['Total'].plot(kind='line')
plt.show()
```

```
plt.figure(figsize=(10,5))
#fig , ax = plt.subplots(figsize=(10,5))
data['Rating'].plot(kind='line')
plt.show()
```

```
data['Gender'].value_counts().plot(kind='pie', autopct='%.2f')
plt.show()
```

# Count Plot

```
sns.countplot(data['Gender'])
plt.show()
```

```
sns.countplot(data=data, x=data['Gender'].value_counts())
plt.show()
```

# HeatMap

```
sns.heatmap(data.corr(numeric_only=True), annot=True)
```

```
sns.heatmap(data.isnull())
plt.show()
```

```
sns.scatterplot(data=data , x='Total',y='cogs')
plt.show()
```

```
plt.scatter(data['Total'],data['cogs'])
plt.show()
```

# Box Plot

```
plt.figure(figsize=(10,5))
sns.boxplot(data=data)
plt.show()
```

```
sns.boxplot(data=data , x = 'Total')
plt.show()
```

```
sns.boxplot(data['cogs'])
plt.show()
```

# Histogram

```
sns.histplot(data=data , x='Total')
plt.show()
```

```
sns.histplot(data['Rating'],bins=50)
plt.show()
```

# Regression Plot

```
sns.regplot(data=data , x='Total', y= 'cogs')
plt.show()
```

```
sns.regplot(data=data , x='Total', y= 'Rating')
plt.show()
```
